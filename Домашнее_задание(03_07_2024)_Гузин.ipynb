### Задание 1: Анализ данных о студентах с использованием NumPy


Описание:
У вас есть данные о студентах, включающие оценки по различным предметам. Используйте библиотеку NumPy для анализа этих данных.

```python
students_data = {
    'math': [85, 78, 92, 70, 88],
    'physics': [90, 82, 76, 85, 89],
    'informatics': [88, 92, 80, 87, 90]
}

students_data
```

Задание:

Найдите среднюю оценку по каждому предмету.<br>
Найдите медианную оценку по каждому предмету.<br>
Вычислите стандартное отклонение по каждому предмету.<br>
Определите предмет с самой высокой средней оценкой.

```python
import numpy as np
student_data_array = np.array(list(students_data.values()))
student_data_array
```

```python
subject_averages = np.mean(student_data_array, axis=1)
print("Средние оценки по предметам:")
print(subject_averages)
```

```python
subject_median = np.median(student_data_array, axis=1)
print("\nМедианные оценки по предметам:")
print(subject_median)
```

```python
subject_std_devs = np.std(student_data_array, axis=1)
print("\nСтандартное отклонение по предметам:")
print(subject_std_devs)
```

```python
highest_avg_subject = np.argmax(subject_averages)
print(f"\nПредмет с самой высокой средней оценкой: {list(students_data.keys())[highest_avg_subject]}")
```

### Задание 2: Анализ данных о продажах с использованием Pandas

Описание:
Используйте библиотеку Pandas для анализа данных о продажах товаров. У вас есть данные о продажах товаров за последний год.

```python
sales_data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'],
    'Sales': [15000, 17000, 16000, 18000, 14000, 19000, 22000, 21000, 23000, 25000, 24000, 26000]
}
sales_data
```

Задание: <br>
Найдите общий объем продаж за год.<br>
Найдите средний объем продаж в месяц.<br>
Найдите месяц с наибольшим и наименьшим объемом продаж.<br>
Постройте линейный график изменения объема продаж по месяцам.

```python
#pip install matplotlib

import matplotlib.pyplot as plt
```

```python
total_sales = np.sum(sales_data['Sales'])
print(f"Общий объем продаж за год: {total_sales}")
```

```python
avg_monthly_sales = np.mean(sales_data['Sales'])
print(f"Средний объем продаж в месяц: {avg_monthly_sales:.2f}")
```

```python
max_sales_month = sales_data['Month'][np.argmax(sales_data['Sales'])]
min_sales_month = sales_data['Month'][np.argmin(sales_data['Sales'])]
print(f"Месяц с наибольшим объемом продаж: {max_sales_month}")
print(f"Месяц с наименьшим объемом продаж: {min_sales_month}")
```

```python
plt.figure(figsize=(12, 6))

plt.plot(sales_data['Month'], sales_data['Sales'])
plt.title('Линейный график')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips') # или другой датасет на выбор

df.head
#проверим что реально некоторых дней недели нет в датасете
monday_data = df[df['day'] == 'Mon']
monday_data.head
```

Постройте следующие графики: <br>
Линейный график <br>
Столбчатый график <br>
Гистограмма <br>
Диаграмма рассеяния <br>

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='day', y='total_bill', data=df)
plt.title('Линейный график общей суммы счета по дням')
plt.xlabel('День')
plt.ylabel('Общая сумма счета')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='day', y='total_bill', data=df)
plt.title('Столбчатый график общей суммы счета по дням')
plt.xlabel('День')
plt.ylabel('Общая сумма счета')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['total_bill'], bins=20)
plt.title('Гистограмма общей суммы счета')
plt.xlabel('Общая сумма счета')
plt.ylabel('Частота')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='day', y='tip', data=df)
plt.title('Диаграмма рассеяния чаевых и общей суммы счета')
plt.xlabel('Общая сумма счета')
plt.ylabel('Чаевые')
plt.show()
```
